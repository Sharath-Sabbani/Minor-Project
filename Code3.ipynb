```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('paddywpi.csv')
df
```

```python
df.Rainfall.median()
```

```python
df.Rainfall = df.Rainfall.fillna(df.Rainfall.median())
df
```

```python
reg = linear_model.LinearRegression() 
reg.fit(df[['Month','Year','Rainfall']],df.WPI)
```

```python
reg.coef_
```

```python
reg.intercept_
```
